MovieLens 1M Dataset

This dataset includes movie ratings data that was collected by GroupLens Research from the users of MovieLens from late 90s to early 00s. Data contains movie ratings, metadata and user demographics.

Dataset contains 1 million ratings from 6000 users on 4000 movies. Data is spread across three tables: ratings, user information, and movie information.

```python
import pandas as pd

# Importing datasets

unames = ["user_id", "gender", "age", "occupation", "zip"]
users = pd.read_table("datasets/users.dat", sep="::", header=None, names=unames, engine="python")

rnames = ["user_id", "movie_id", "rating", "timestamp"]
ratings = pd.read_table("datasets/ratings.dat", sep="::", header=None, names=rnames, engine="python")

mnames = ["movie_id", "title", "genres"]
movies = pd.read_table("datasets/movies.dat", sep="::", header=None, names=mnames, engine="python")
```

```python
# Checking data sets

users.head(5)
```

```python
ratings.head(5)
```

```python
movies.head(5)
```

```python
# Merging ratings, users, and movies table

data = pd.merge(pd.merge(ratings, users), movies)

data
```

```python
# Mean movie ratings for each film grouped by gender

mean_ratings_by_gender = data.pivot_table("rating", index="title", columns="gender", aggfunc="mean")

mean_ratings_by_gender.head(5)
```

```python
# Index of movie titles that at least got 250 ratings

ratings_by_title = data.groupby("title").size()

ratings_by_title.head()
```

```python
active_titles = ratings_by_title.index[ratings_by_title >= 250]

active_titles
```

```python
# This index is used to select titles from mean_ratings_by_gender table

mean_ratings_by_gender = mean_ratings_by_gender.loc[active_titles]

mean_ratings_by_gender
```

```python
# Top 10 highest rated movie by women

top_female_ratings = mean_ratings_by_gender.sort_values("F", ascending=False)

top_female_ratings.head(10)
```

```python
# Top 10 highest rated movie by men

top_male_ratings = mean_ratings_by_gender.sort_values("M", ascending=False)

top_male_ratings.head(10)
```

```python
# Most devisive movies betweeen genders

mean_ratings_by_gender["diff"] = mean_ratings_by_gender["M"] - mean_ratings_by_gender["F"]
```

```python
# Most devisive movies that women preferred

sorted_by_diff = mean_ratings_by_gender.sort_values("diff")

sorted_by_diff.head(5)
```

```python
# Most devisive movies that men preferred

sorted_by_diff[::-1].head(5)
```

```python
# Most devisive movies irregardless of gender (by calculating the standart deviation)

rating_std_by_title = data.groupby("title")["rating"].std()

rating_std_by_title = rating_std_by_title.loc[active_titles]

rating_std_by_title.sort_values(ascending=False)[:10]
```

```python
# Exploring genres in the dataset

movies["genres"].head()
```

```python
# Splitting the single string of genres into multiple elements by category

movies["genres"].head().str.split("|")
```

```python
movies["genre"] = movies.pop("genres").str.split("|")

movies.head()
```

```python
# Creating a new dataframe in which each inner genre element corresponds to one row

movies_exploded = movies.explode("genre")

movies_exploded.head(10)
```

```python
# Merging all tables using group by genre

ratings_with_genre = pd.merge(pd.merge(movies_exploded, ratings), users)

ratings_with_genre.iloc[0]
```

```python
# Movie ratings by genre and age group

genre_ratings = (ratings_with_genre.groupby(["genre", "age"])["rating"].mean().unstack("age"))

genre_ratings[:10]
```

```python
# Average rating of genres by gender

mean_genre_ratings_per_gender = ratings_with_genre.groupby(["genre", "gender"])["rating"].mean()

mean_genre_ratings_per_gender.head()
```

```python
mean_genre_ratings_per_gender = mean_genre_ratings_per_gender.unstack("gender")

mean_genre_ratings_per_gender
```

```python
# Top 5 genres that have been highly rated by men

mean_genre_ratings_per_gender.sort_values("M", ascending=False).head(5)
```

```python
# Top 5 genres that have been highly rated by women

mean_genre_ratings_per_gender.sort_values("F", ascending=False).head(5)
```

```python
# Differences in rating by gender
# Genres with negative difference values are higher rated by men than women
# Genres with positive difference values are higher rated by women than men

mean_genre_ratings_per_gender["diff"] = mean_genre_ratings_per_gender["F"] - mean_genre_ratings_per_gender["M"]

mean_genre_ratings_per_gender.sort_values("diff")
```

```python
# Most devisive genres betweeen genders

mean_genre_ratings_per_gender["abs(diff)"] = abs(mean_genre_ratings_per_gender["diff"])

mean_genre_ratings_per_gender.sort_values("abs(diff)", ascending=False).head(5)
```
